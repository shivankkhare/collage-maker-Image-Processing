```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = cv2.imread('top_left.jpg')
```

```python
cv2.imshow('sample image',img1)
cv2.waitKey(0) # waits until a key is pressed
cv2.destroyAllWindows() # destroys the window showing image
```

```python
img1.shape
```

```python
new_img1 = cv2.resize(img1,(200,200))
```

```python
cv2.imshow('sample image',new_img1)
cv2.waitKey(0) # waits until a key is pressed
cv2.destroyAllWindows() # destroys the window showing image
```

```python
new_img1 =cv2.copyMakeBorder(new_img1, 0, 5, 0, 5, cv2.BORDER_CONSTANT)
```

```python
cv2.imshow('sample image',new_img1)
cv2.waitKey(0) # waits until a key is pressed
cv2.destroyAllWindows() # destroys the window showing image
```

```python
img2 =  cv2.imread('top_right.jpg')
new_img2=cv2.resize(img2,(200,200)) 
new_img2 =cv2.copyMakeBorder(new_img2, 0, 5, 5, 0, cv2.BORDER_CONSTANT)
plt.imshow(new_img2)
```

```python
im = cv2.hconcat([new_img1,new_img2])
plt.imshow(im)
```

```python
img3 =  cv2.imread('bottom_left.jpg')
new_img3=cv2.resize(img3,(200,200)) 
new_img3 =cv2.copyMakeBorder(new_img3, 5, 0, 0, 5, cv2.BORDER_CONSTANT)
plt.imshow(new_img3)
```

```python
img4 =  cv2.imread('bottom_right.jpg')
new_img4=cv2.resize(img4,(200,200)) 
new_img4 =cv2.copyMakeBorder(new_img4, 5, 0, 5, 0, cv2.BORDER_CONSTANT)
plt.imshow(new_img4)
```

```python
im2 = cv2.hconcat([new_img3,new_img4])
plt.imshow(im2)
```

```python
im = cv2.vconcat([im,im2])
plt.style.use('seaborn')
plt.imshow(im)
```

```python
fields = ['r', 'g', 'b'] 
with open('pokemon3.csv', 'w', newline='') as file:
    # creating a csv writer object  
    csvwriter = csv.writer(file)  
    csvwriter.writerow(fields) 
    for i in new_im:
        csvwriter.writerows(i) 
```
